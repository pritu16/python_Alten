```python
#1
import pandas as pd
import numpy as np
data={'Id':[1,2,3,4,5,6,None],
     'Name':['Shivansh',None,'Swati',None,'Arpit','Ritik',None],
     'Job Profile':['Data Scientist','Software Developer','Data Analyst',None,'Android Developer',None,None],
     'City':['Noida',None,'Hyderabad','Noida','Bangalore',None,None]}
indexes=['r1','r2','r3','r4','r5','r6','r7']
df=pd.DataFrame(data,index=indexes)
df
```

```python
#i) Cleaning the Data
df.drop(['r7'],inplace=True)
df
```

```python
df.drop(['r2'],inplace=True)
df
```

```python
#ii) new column
height=[5.1,6.2,None,5.1,5.2]
df['Height']=height
df
```

```python
#column mean
df1=df['Height'].mean()
df1
```

```python
#replacing the height column none value with mean value
df['Height'] = df['Height'].fillna(5.3999999999999995)
df
```

```python
#fill the data with default city Bangalore
df['City']=df['City'].fillna('Bangalore')
df
```

```python
#2
import pandas as pd
import numpy as np
raw_Data={'Emp_Name':['Alten1','Alten2','Alten3','Alten4','Alten5','Alten6','Alten7','Alten8'],
         'Tenure':[1,3,5,9,7,4,2,None]}
df2=pd.DataFrame(raw_Data)
df2
```

```python
#new column 'promotion' if the tenure is >5
promotion=[]
for row in df2['Tenure']:
    if row>5:
        promotion.append('Yes')
    else:
        promotion.append('No')
df2['Promotion']=promotion
df2
```

```python
#
df3=df2['Tenure'].mean()
df3
```
